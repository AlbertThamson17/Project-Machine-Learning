```python
import numpy as np
import pandas as pd
import math
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler , OrdinalEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('responses.csv')
df = df.dropna()
print(df.isna().values.any())
df
```

```python
#Smoking  
df["Smoking"] = df["Smoking"].replace('current smoker', 3)  
df["Smoking"] = df["Smoking"].replace('former smoker', 2) 
df["Smoking"] = df["Smoking"].replace('tried smoking', 1)   
df["Smoking"] = df["Smoking"].replace('never smoked', 0)
#Alcohol
df["Alcohol"] = df["Alcohol"].replace('drink a lot', 2)  
df["Alcohol"] = df["Alcohol"].replace('social drinker', 1)    
df["Alcohol"] = df["Alcohol"].replace('never', 0)
```

```python
feature = df[["Music","Slow songs or fast songs","Dance","Folk","Country","Classical music","Musical","Pop","Rock","Metal or Hardrock","Punk","Hiphop, Rap","Reggae, Ska","Swing, Jazz","Rock n roll","Alternative","Latino","Techno, Trance","Opera"]]
target = df[["Smoking","Alcohol","Healthy eating"]]
```

```python
#Normalize dataset
from sklearn import preprocessing
#print(dataset.shape)
x = feature
y = target
x = preprocessing.normalize(x)
#print(normalize_dataset)
```

```python
#Analisa dengan PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=11)
principalComponent = pca.fit_transform(x)

principalA = pd.DataFrame(data = principalComponent, columns=['principal component 1', 'principal component 2', 'principal component 3', 'principal component 4','principal component 5', 'principal component 6', 'principal component 7', 'principal component 8', 'principal component 9', 'principal component 10', 'principal component 11'])
finalA = pd.concat([principalA, target['Smoking']] , axis = 1)
print(finalA)
```

```python
finalA = finalA.dropna()
print(finalA.isna().values.any())
```

```python
#import tf1
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

#load dataset yg sudah d PCA
def load_dataset():
    feature = finalA[['principal component 1', 'principal component 2', 'principal component 3', 'principal component 4','principal component 5', 'principal component 6', 'principal component 7', 'principal component 8', 'principal component 9', 'principal component 10', 'principal component 11']]
    target = finalA[['Smoking']]
    return feature, target

feature, target = load_dataset()

#generate layer,weight and bias
layer = {
    'input' : 11,
    'hidden' : 3,
    'output' : 4
}

weight = {
    'input_to_hidden' : tf.Variable(tf.random_normal([layer['input'], layer['hidden']])),
    'hidden_to_output' : tf.Variable(tf.random_normal([layer['hidden'], layer['output']])) 
}
bias = {
    'input_to_hidden' : tf.Variable(tf.random_normal([layer['hidden']])),
    'hidden_to_output' : tf.Variable(tf.random_normal([layer['output']]))
}

feature = OrdinalEncoder().fit_transform(feature)
feature = MinMaxScaler().fit_transform(feature)

target = OneHotEncoder(sparse=False).fit_transform(target)

input_placeholder = tf.placeholder(tf.float32, [None, layer['input']])
output_placeholder = tf.placeholder(tf.float32, [None, layer['output']])

def feed_forward(dataset):
    weightxbias1 = tf.matmul(dataset, weight['input_to_hidden']) + bias['input_to_hidden']
    activation1 = tf.nn.sigmoid(weightxbias1)

    weightxbias2 = tf.matmul(activation1, weight['hidden_to_output']) + bias['hidden_to_output']
    activation2 = tf.nn.sigmoid(weightxbias2)

    return activation2
#generate learning rate and epoch
learning_rate = 0.3
epoch = 5000
#train dataset using mean squared error and gradient descent optimizer
output = feed_forward(input_placeholder)
mse = tf.reduce_mean(0.5 * (output - output_placeholder) ** 2) 
optimizer = tf.train.GradientDescentOptimizer(learning_rate)
train = optimizer.minimize(mse)
saver = tf.train.Saver()
feature_train, feature_test, target_train, target_test = train_test_split(feature,target,test_size=0.2)

#data training
with tf.Session() as sess :
    sess.run(tf.global_variables_initializer())

    feed_train = {
        input_placeholder: feature_train,
        output_placeholder: target_train
    }
    
    for i in range(epoch):
        sess.run(train, feed_dict=feed_train)
        error = sess.run(mse, feed_dict = feed_train)
        if i % 100 == 0:
            print('Epoch: {}, Current Error: {}'. format(i+100, error))
        
                       
    feed_test = {
        input_placeholder: feature_test,
        output_placeholder: target_test
    }
    
    #printing epohc, error and accuracy 
    matches = tf.equal(tf.argmax(output, axis=1),tf.argmax(output_placeholder,axis=1))
    accuracy = tf.reduce_mean(tf.cast(matches,tf.float32))
    print('Accuracy: {}%'.format(sess.run(accuracy*100, feed_dict =feed_test)))
```
